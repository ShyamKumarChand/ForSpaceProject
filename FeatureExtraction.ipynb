```python
from sklearn.feature_extraction.text import CountVectorizer
text = ["The quick brown fox jumped over the lazy dog,The King"]
```

```python
Vectorizer = CountVectorizer()
Vectorizer.fit(text)
```

```python
print(Vectorizer.vocabulary_)
```

```python
vector = Vectorizer.transform(text)
```

```python
print(vector.shape)
print(type(vector))
print(vector.toarray())
```

```python
text2 = ["the puppy"]
vector = Vectorizer.transform(text2)
print(vector.toarray())
```

```python
Sample_text = ["One of the most basic ways we can numerically represents words "
              "is through the one-hot encodeing method(also sometimes called (countVectorizer)"]
```

```python
vector = Vectorizer.transform(Sample_text)
```

```python
print('Full vector: ')
print(vector.toarray())
```

```python
print('Hot Vector:')
print(Vectorizer.transform(['hot']).toarray())
```

```python
print('Hot and one:')
print(Vectorizer.transform(['hot', 'one']).toarray())
```

```python
print('One swoop:')
new_text = ['Today is the day that I do the thing today, today']
new_vectorizer = CountVectorizer()
print(new_vectorizer.fit_transform(new_text).toarray())
```
