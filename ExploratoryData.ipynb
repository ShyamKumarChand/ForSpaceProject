```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pandas_profiling as pp
```

```python
URL_AcidQuality = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
```

```python
AcidQuality = pd.read_csv(URL_AcidQuality,sep=';')
```

```python
bin_edges = np.arange(0, AcidQuality['residual sugar'].max() + 1, 1)
plt.figure(figsize=(20,9))
fig = plt.hist(AcidQuality['residual sugar'], bins=bin_edges)

# add plot labels
plt.xlabel('count')
plt.ylabel('residual sugar')
plt.show()
```

```python
plt.figure(figsize=(20,9))
fig = plt.scatter(AcidQuality['pH'], AcidQuality['residual sugar'])

# add plot labels
plt.xlabel('pH')
plt.ylabel('residual sugar')
plt.show()
```

```python
plt.figure(figsize=(20,9))
fig = sns.pairplot(data=AcidQuality[['alcohol', 'pH', 'residual sugar', 'quality']], 
                   hue='quality')

# add plot labels
plt.xlabel('pH')
plt.ylabel('residual sugar')
plt.show()
```
