```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80,1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 -rng.rand(16)) 
```

```python
reg_1 = DecisionTreeRegressor(max_depth=2)
reg_1.fit(X,y)
```

```python
reg_1 = DecisionTreeRegressor(max_depth=5)
reg_1.fit(X,y)
```

```python
X_test = np.arange(0.0,5.0,0.01)[:,np.newaxis]
y_1 = reg_1.predict(X_test)
```

```python
X_test = np.arange(0.0,5.0,0.01)[:,np.newaxis]
y_2 = reg_1.predict(X_test)
```

```python
Train_score = reg_1.score(X,y)
```

```python
Train_score
```

```python
Test_score = reg_1.score(X_test,y_2)
```

```python
Test_score
```

```python
plt.figure(figsize=(30,16))
plt.scatter(X,y,s = 20, edgecolor="black",
           c='darkorange',label='data')
plt.plot(X_test,y_1,color='cornflowerblue',
        label='max_depth=2',linewidth=2)
plt.plot(X_test,y_2,color='yellowgreen',
       label='max_depth=5',linewidth=5)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regressor")
plt.legend()
plt.show()
```

```python
rng = np.random.RandomState(1)
X = np.sort(200 * rng.rand(100,1), axis=0)
y = np.array([np.pi*np.sin(X).ravel(),np.pi*np.cos(X).ravel()]).T
y[::5,:] += 3 * (0.5 -rng.rand(20, 2)) 
```

```python
reg_1 = DecisionTreeRegressor(max_depth=2)
reg_2 = DecisionTreeRegressor(max_depth=5)
reg_3 = DecisionTreeRegressor(max_depth=8)
reg_1.fit(X,y)
reg_2.fit(X,y)
reg_3.fit(X,y)
```

```python
X_test = np.arange(-100.0,100.0,0.01)[:,np.newaxis]
y_1 = reg_1.predict(X_test)
y_2 = reg_2.predict(X_test)
y_3 = reg_3.predict(X_test)
```

```python
plt.figure(figsize=(30,16))
s = 140
plt.scatter(y[:, 0],
            y[:, 1],
            c="navy", 
            s=s,
            edgecolor="black", 
            label="data")
plt.scatter(y_1[:, 0],
            y_1[:, 1],
            c="cornflowerblue", 
            s=s,
            edgecolor="black", 
            label="max_depth=2")
plt.scatter(y_2[:, 0],
            y_2[:, 1],
            c="red", 
            s=s,
            edgecolor="black", 
            label="max_depth=5")
plt.scatter(y_3[:, 0], 
            y_3[:, 1], 
            c="orange", 
            s=s,
            edgecolor="black", 
            label="max_depth=8")
```
