```python
import numpy as np
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
import statistics as stat
import pandas as pd
```

```python
URL = 'https://raw.githubusercontent.com/reddyprasade/Machine-Learning-Problems-DataSets/master/Classification/mushrooms.csv'
```

```python
Data = pd.read_csv(URL)
Data.head()
```

```python
Data.tail()
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data_encode = pd.get_dummies(Data)
Data_encode
```

```python
X_mush = Data_encode.iloc[:,2:]
y_mush = Data_encode.iloc[:,1]
X_mush
```

```python
y_mush
```

```python
X_train,X_test, y_train, y_test=train_test_split(X_mush,y_mush,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model1 = DecisionTreeClassifier(max_depth=4)
model2 = KNeighborsClassifier(n_neighbors=7)
model3 = SVC(kernel='rbf',probability=True)
```

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
```

```python
Score1 = model1.score(X_test,y_test)
Score1
```

```python
Score2 = model2.score(X_test,y_test)
Score2
```

```python
Score3 = model3.score(X_test,y_test)
Score3
```

```python
pred1 = model1.predict(X_test)
pred2 = model2.predict(X_test)
pred3 = model3.predict(X_test)
```

```python
final_pred = np.array([])
for i in range(0,len(X_test)):
    final_pred = np.append(final_pred,stat.mode([pred1[i],
                                                pred2[i],
                                                pred3[i]]))
final_pred    
```

```python
total = np.sum(y_test==final_pred)
print("Accuracy:",total,"/",len(final_pred),"*100=","{0:.3f}".format(total/len(final_pred)*100),"%")
```

```python
spred1 = model1.predict_proba(X_test)
spred2 = model2.predict_proba(X_test)
spred3 = model3.predict_proba(X_test)
```

```python
finalpred= (spred1+spred2+spred3)/3
finalpred
```

```python
final_classes = finalpred.argmax(axis=-1)
final_classes
```

```python
total = np.sum([y_test == final_classes[i] for i in range(len(final_classes))])
print("Accuracy:", total,"/",len(final_classes),"*100=","{0:.3f}".format(total/len(final_classes)*100),"%")
```
